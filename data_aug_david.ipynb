```python
#https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html

from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

img = load_img('/content/IMG_20240924_185006.jpg')  # this is a PIL image

x = img_to_array(img)  # this is a Numpy array with shape (3, 150, 150)
input_batch = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 3, 150, 150)
input_batch.shape
```

```python
# the .flow() command below generates batches of randomly transformed images
# and saves the results to the `preview/` directory

import os

# Create the 'preview' directory if it doesn't exist
if not os.path.exists('preview'):
    os.makedirs('preview')

i = 0
for batch in datagen.flow(input_batch, batch_size=1,
                          save_to_dir='preview', save_prefix='dog'):
    i += 1
    if i > 8:
        break  # otherwise the generator would loop indefinitely
```
